```python
import streamlit as st
import psycopg2
from datetime import datetime

# ------------------ PostgreSQL Connection ------------------
db_name = "DDSA_NEW"
table_name = "feedback"
host = "localhost"
port = "5432"
user = "postgres"
password = "123"

conn = psycopg2.connect(
    host=host,
    database=db_name,
    user=user,
    password=password,
    port=port
)
cur = conn.cursor()

# ------------------ Create table if not exists ------------------
cur.execute(f"""
CREATE TABLE IF NOT EXISTS {table_name} (
    id SERIAL PRIMARY KEY,
    action VARCHAR(20),
    rating INTEGER,
    comment TEXT,
    created_at TIMESTAMP DEFAULT NOW()
);
""")
conn.commit()

# ------------------ Footer with Ribbon + Name ------------------
st.markdown("""
<div style="
    text-align: center; 
    color: #111827; 
    font-size: 22px;  
    font-weight: 700;
    margin-top: 30px;
    padding: 10px 0;
    position: relative;
">
    <div style="
        width: 140px; 
        height: 4px; 
        background: linear-gradient(90deg, #2563EB, #10B981); 
        margin: 0 auto 12px auto;
        border-radius: 2px;
    "></div>
    Created by Sholingan
</div>
""", unsafe_allow_html=True)

# ------------------ Like/Dislike + Star Rating ------------------
col_left, col_right = st.columns([1,1])

# --- Left Column: Like/Dislike ---
with col_left:
    if st.button("üëç Like"):
        cur.execute(f"INSERT INTO {table_name} (action, created_at) VALUES (%s, %s)", ("Like", datetime.now()))
        conn.commit()
        st.success("Thanks for your feedback! üëç")
    if st.button("üëé Dislike"):
        cur.execute(f"INSERT INTO {table_name} (action, created_at) VALUES (%s, %s)", ("Dislike", datetime.now()))
        conn.commit()
        st.warning("Thanks for your honesty! We'll improve.")

# --- Right Column: Star Rating ---
with col_right:
    st.markdown("### ‚≠ê Rate this Dashboard")
    rating = st.radio("", options=[1,2,3,4,5], format_func=lambda x: "‚≠ê"*x)
    if st.button("Submit Rating", key="rating_submit"):
        cur.execute(f"INSERT INTO {table_name} (rating, created_at) VALUES (%s, %s)", (rating, datetime.now()))
        conn.commit()
        st.success(f"Thanks for rating us {rating} star{'s' if rating>1 else ''}! üåü")

# ------------------ Comment Section ------------------
st.markdown("""
<div style="text-align:center; margin-top:20px; font-family: 'Roboto', sans-serif;">
    <div style="margin-bottom:10px; font-weight:600; color:#111827; font-size:16px;">
        Share your thoughts or share this dashboard:
    </div>
</div>
""", unsafe_allow_html=True)

comment = st.text_area("üí¨ Leave a comment", placeholder="Type your feedback here...", key="user_comment")

col1, col2, col3 = st.columns([1,1,1])

# Share / Submit
with col1:
    if st.button("üîó Share on Twitter"):
        st.write("You can share this dashboard on Twitter!")  # Add actual link

with col2:
    if st.button("üîó Share on LinkedIn"):
        st.write("You can share this dashboard on LinkedIn!")  # Add actual link

with col3:
    if st.button("Submit Comment"):
        if comment.strip() != "":
            cur.execute(f"INSERT INTO {table_name} (comment, created_at) VALUES (%s, %s)", (comment, datetime.now()))
            conn.commit()
            st.success("Thanks for your comment! ‚úÖ")
        else:
            st.warning("Please type a comment before submitting.")

# ------------------ Close Connection ------------------
cur.close()
conn.close()
```
