```python
import streamlit as st
import psycopg2
import pandas as pd

# ---------------- PostgreSQL Connection ----------------
db_name = "DDSA_NEW"
table_name = "dashboard_feedback"
host = "localhost"
port = "5432"
user = "postgres"
password = "123"

def get_connection():
    return psycopg2.connect(
        dbname=db_name, user=user, password=password, host=host, port=port
    )

# ---------------- Insert Feedback ----------------
def insert_feedback(feedback_type, rating=None, comment=None):
    conn = get_connection()
    cur = conn.cursor()
    cur.execute(
        f"INSERT INTO {table_name} (type, value, comment) VALUES (%s, %s, %s)",
        (feedback_type, rating, comment)
    )
    conn.commit()
    cur.close()
    conn.close()

# ---------------- Fetch Feedback Counts ----------------
def get_feedback_counts():
    conn = get_connection()
    cur = conn.cursor()
    cur.execute(f"""
        SELECT type, value, COUNT(*) 
        FROM {table_name}
        GROUP BY type, value
    """)
    rows = cur.fetchall()
    cur.close()
    conn.close()
    return rows
```
