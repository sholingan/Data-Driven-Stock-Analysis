```python
import pandas as pd

# Load stock data
df = pd.read_csv(r"D:\Py_start\Python\project_SN\DDSA\new\df2.csv")

print(df.head())
```

```python
# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Pivot table
pivot_df = df.pivot(index='date', columns='Ticker', values='close')

print(pivot_df.head())
```

```python
# Calculate correlation
corr_matrix = pivot_df.corr()

print(corr_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,16))  # Increase figure size

sns.heatmap(corr_matrix, 
            annot=True,       # Remove numbers (optional for cleaner look)
            cmap='coolwarm',
            annot_kws={"size": 6}, 
            linewidths=0.5,
            cbar=True,         # Show colorbar
            square=True)

plt.title('Stock Price Correlation Heatmap', fontsize=20)
plt.xticks(rotation=90)  # Rotate X-axis labels
plt.yticks(rotation=0)   # Y-axis labels straight
plt.tight_layout()
plt.show()
```

```python
corr_matrix.to_csv(r"D:\Py_start\Python\project_SN\DDSA\new\stock_correlation_matrix.csv")
```
